```python
import numpy as np
import pandas as pd
import os

DATEID = "20250308"  # Example date ID
output_dir = "store_data"  # Directory to store the generated files

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

for i in range(1, 101):  # Loop through StoreID 1 to 100
    num_rows = np.random.randint(100, 1000)  # Random number of rows

    print(f"ðŸ”¹ Processing StoreID {i} with {num_rows} rows...")  

    data = {
        'DateID': np.random.randint(20230101, 20231231, size=num_rows),  # Example DateID
        'ProductID': np.random.randint(1, 1001, size=num_rows),
        'StoreID': [i] * num_rows,  
        'CustomerID': np.random.randint(1, 1001, size=num_rows),
        'QuantityOrdered': np.random.randint(1, 21, size=num_rows),
        'OrderAmount': np.random.randint(100, 1001, size=num_rows),
        'ShippingCost': np.random.uniform(5, 50, size=num_rows).round(2),
        'DiscountAmount': np.random.uniform(0, 100, size=num_rows).round(2),
        'TotalAmount': lambda df: df['OrderAmount'] + df['ShippingCost'] - df['DiscountAmount']
    }

    df = pd.DataFrame(data)
    df['TotalAmount'] = df['OrderAmount'] + df['ShippingCost'] - df['DiscountAmount']  # Compute TotalAmount
    
    # Define the file path for each StoreID
    file_path = os.path.join(output_dir, f"store_{i}.csv")
    
    # Save to CSV
    df.to_csv(file_path, index=False)

print("âœ… Data generation complete. Check the 'store_data' folder for output files.")
```
